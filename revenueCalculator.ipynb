```python
def revenueCalculator(actual, predicted):
    revenue = 0
    biding_on = []
    i = 0
    num_of_biding = 0
    while revenue <= 10000 and i<len(actual):
        if num_of_biding >= 4000:
            break
        if(predicted[i]>=actual[i] and predicted[i] < 2*actual[i]):
            biding_on.append(i)
            revenue = revenue + 2*predicted[i] - actual[i]
            num_of_biding += 1
        i = i + 1
    
    return (revenue, biding_on)
```

```python
revenueCalculator([2,3],[3,4])
```
